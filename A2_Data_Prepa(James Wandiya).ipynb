# **Tourism Analysis at The Amazon Monument in Benin**  

### **Background**  
The **Amazon Monument** in Benin, inaugurated on **July 30, 2022**, honors the legendary Amazons of the Kingdom of Dahomey. Similar to the **Monument de la Renaissance Africaine** in Dakar, it celebrates **African heritage, pride, resilience**, and symbolizes the **strength and historical legacy** of the people.

Since its inauguration, the monument has become a popular destination, attracting visitors from all over the world. To monitor tourism trends, the **Agence Nationale du Tourisme Bninois** has started recording data **every three days** on various aspects of visitor activity at the monument.  It is stored in the `data` folder as `Tourist_Visits_at_Amazon_Monument.csv`

Your task is to **analyze the collected data**  to uncover trends and patterns that can help the tourism agency improve visitor experiences and manage operations better.  


### **Dataset Overview**

The dataset provided contains the following columns:

1. **Date**:  
   - The date when the data was recorded (every 3 days).  

2. **Number of Adults**:  
   - The total number of adult visitors (18 years and above).  

3. **Number of Teenagers**:  
   - The total number of teenage visitors (13-17 years old).  

4. **Number of Children**:  
   - The total number of child visitors (under 13 years).  

5. **Local vs Foreign Visitors Ratio**:  
   - The proportion of local (Beninese) visitors compared to foreign tourists (value between 0 and 1).  


6. **Daily Revenue (USD)**:  
   - The total revenue generated from ticket sales and other activities on that day.  

7. **Weather Conditions**:  
   - The weather on the recorded day (Sunny, Rainy, Cloudy).  

8. **Temperature (C)**:  
   - The average temperature on the recorded day.  
 

10. **Photo-Taking Frequency**:  
   - The number of photos taken or social media check-ins on that day, reflecting visitor engagement.
