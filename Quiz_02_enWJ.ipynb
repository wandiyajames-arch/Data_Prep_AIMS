In Cameroon’s rainforests, there are two amazing animals: the **Goliath frog** and the **African grey parrot**.

- The **African grey parrot** is a very smart bird. It can live for **50 years** and is usually **gray** with a **red tail**. It can also copy human speech!


### **Question 1: Identify Attributes**
**Task:** For each animal, list at least **three attributes** that could be used to create a `Frog` and a `Parrot` dataclass. Use the description above to guide you.


```python
from dataclasses import dataclass

@dataclass
class Frog:
    species: str
    max_length_cm: float
    max_weight_kg: float
    habitat: str
    skin_color: str
    conservation_status: str

@dataclass
class Parrot:
    species: str
    lifespan_years: int
    primary_color: str
    accent_color: str
    can_mimic_speech: bool
    intelligence_level: str

goliath_frog = Frog(
    species="Goliath frog",
    max_length_cm=32.0,
    max_weight_kg=3.3,
    habitat="rainforest rivers",
    skin_color="greenish-brown",
    conservation_status="Endangered"
)

african_grey = Parrot(
    species="African grey parrot",
    lifespan_years=50,
    primary_color="gray",
    accent_color="red",
    can_mimic_speech=True,
    intelligence_level="very high"
)
```

```python
print(african_grey)
```

### **Question 2: Create Dataclasses**

 Create two subclasses: `Frog` and `Parrot`, each with their own unique attributes (e.g., `Frog` could have `length: float`, `weight: float`, and `habitat: str`). Instantiate one object for each subclass.


```python
from dataclasses import dataclass
from typing import Optional

@dataclass
class Animal:
    species: str
    species_status: str
    region: str


@dataclass
class Frog(Animal):
    length_cm: float
    weight_kg: float
    habitat: str
    complexion: str
    is_poisonous: bool = False  

 
@dataclass
class Parrot(Animal):
    years: int
    color1: str
    color2: Optional[str] = None
    talking: bool = False
    height_cm: Optional[float] = None
```

**Task:** Modify the `Parrot` dataclass so that the `color` attribute can only be one of the following: `"gray"`, `"red"`, or `"white"`. Use Python’s `Literal` type to enforce this.


```python
from dataclasses import dataclass
from typing import Optional
from typing import Literal

@dataclass
class Animal:
    species: str
    conservation_status: str
    native_region: str


@dataclass
class Frog(Animal):
    length_cm: float
    weight_kg: float
    habitat: str
    skin_color: str
    is_poisonous: bool = False


@dataclass
class Parrot(Animal):
    lifespan_years: int
    primary_color: Literal["gray", "red", "white"]
    accent_color: Optional[Literal["gray", "red", "white"]] = None
    can_mimic_speech: bool = False
    wingspan_cm: Optional[float] = None




```
