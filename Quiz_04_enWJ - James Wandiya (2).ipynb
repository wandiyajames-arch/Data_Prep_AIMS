## 3. Run the command below to generate a dataframe

```python
import numpy as np
my_initial_df = pd.DataFrame({'year': list(np.random.randint(2000,2025,10)), 
                   'month': list(np.random.randint(1,13,10)), 
                   'day': list(np.random.randint(1,29,10)), 
                   'hour': list(np.random.randint(0,24,10)), 
                   'meal_menu': ['mafe', 'yassa', 'cebon', 'soup', 'mafe', 'grilled fish', 'grilled fish', 'yassa', 'cebon', 'soup']})
my_initial_df['date'] = pd.to_datetime(my_initial_df[['year', 'month', 'day', 'hour']])


my_initial_df  = my_initial_df.drop(columns=['year', 'month', 'day', 'hour'])
my_initial_df = my_initial_df[['date', 'meal_menu']]
my_initial_df
```

```python
print(unique_meals)
```

## 9. Create a Timedelta object representing:  `5 days, 3 hours, and 30 minutes` and add `two weeks 10 minutes` that. 



**<span style="color:red">Score:  1/ 1 </span>**

```python
td1 = pd.Timedelta(days=5, hours=3, minutes=30)
td2 = pd.Timedelta(weeks=2, minutes=10)
total = td1 + td2
```

```python
print("Timedelta 1:", td1)
```

```python
print("Timedelta 2:", td2)

```

```python
print("Total:", total)
```
